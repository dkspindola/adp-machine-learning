```python
# Plots für die gesamten Bereiche für die Rohdaten
from Read_Data_SingleBlech import Read_Data
import matplotlib.pyplot as plt
import os 

# Durch Angabe der Range können die Bleche die geplottet werden sollen ausgewählt werden

for i in range(98,101):
    
    Daten, BlechNummer = Read_Data(i)
    
    # Abfrage notwendig, falls eine Blech Nummer im Ordner nicht existiert wird in der FUnktion Read_Data_SingleBlech None ausgegeben,so wird dieses Blech einfach übersprungen
    if Daten is not None:
        # Festlegung des Sicherungsordners, muss entsprechend angepasst werden
        Sicherungsordner = f'C:\\Users\\corvi\\OneDrive - stud.tu-darmstadt.de\\Desktop\\Masterthesis\\15_Plots\\Plots_{BlechNummer}'
        # Falls Sicherungsordner noch nicht existiert, wird dieser automatisch erstellt
        if os.path.exists(Sicherungsordner):
            print(f'{Sicherungsordner} --> Existiert bereits')
        else:
            os.makedirs(Sicherungsordner)
            
        #Berechnung des Moving Averages der Sensoren
        MovingAverage = Daten.rolling(window=5).mean()
        
        # Unterscheidung zwischen Lateralen und Axialen Kräfte, sodass später bei den Plots einzeln darauf zugegriffen werden kann
        Lateral = [col for col in Daten.columns if 'Lateral' in col]
        Axial = [col for col in Daten.columns if 'Axial' in col]
        
        # Plots aller Sensoren eines Bleches in einem Graphen kann ggfs. angepasst werden
        Daten.plot( linestyle='-', use_index=True, figsize=(10,6))
        plt.title(f'Blech {BlechNummer} Gesamtübersicht mit allen Sensoren', fontsize=15, pad=12)
        plt.xlabel('Datenpunkte', fontsize=14, labelpad=12)
        plt.ylabel('Kraft in [N]', fontsize=14,labelpad=12)
        plt.legend(loc='upper right', fontsize=11)
        plt.grid(True)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.tight_layout()
        #plt.show()
        plt.savefig(os.path.join(Sicherungsordner, f'{BlechNummer}_Plot_Gesamteübersicht von allen Sensoren.svg'), format='svg')
        plt.savefig(os.path.join(Sicherungsordner, f'{BlechNummer}_Plot_Gesamteübersicht von allen Sensoren.png'), format='png')
        plt.close()

        #print(MovingAverage_Axial.iloc[50:100])
        #print(Daten[Axial].iloc[50:100])
        #print(Lateral)

        # For-Schleife um alle Lateralen und Axialen Kräfte in einem Plot zu plotten und die oben berechneten MovingAverages ebenfalls 
        for Kraft, Liste in [('Laterale', Lateral), ('Axiale', Axial)]:
            Daten[Liste].plot( linestyle='-', use_index=True, figsize=(10,6))
            plt.title(f'Blech {BlechNummer} {Kraft} Kräfte', fontsize=19, pad=12)
            plt.xlabel('Datenpunkte', fontsize=18,labelpad=12)
            plt.ylabel('Kraft in [N]', fontsize=18,labelpad=12)
            plt.legend(loc='upper right', fontsize=14)
            plt.grid(True)
            plt.xticks(fontsize=15)
            plt.yticks(fontsize=15)
            #plt.show()
            plt.tight_layout()
            plt.savefig(os.path.join(Sicherungsordner, f'{BlechNummer}_Plot_{Kraft}Kräfte.svg'), format='svg')
            plt.savefig(os.path.join(Sicherungsordner, f'{BlechNummer}_Plot_{Kraft}Kräfte.png'), format='png')
            plt.close()

            MovingAverage[Liste].plot( linestyle='-', use_index=True, figsize=(10,6))
            plt.title(f'Blech {BlechNummer} {Kraft} Kräfte Moving Average', fontsize=19,pad=12)
            plt.xlabel('Datenpunkte', fontsize=18,labelpad=12)
            plt.ylabel('Kraft in [N]', fontsize=18,labelpad=12)
            plt.legend(loc='upper right', fontsize=14)
            plt.grid(True)
            plt.xticks(fontsize=15)
            plt.yticks(fontsize=15)
            #plt.show()
            plt.tight_layout()
            plt.savefig(os.path.join(Sicherungsordner, f'{BlechNummer}_Plot_{Kraft} Kräfte Moving_Average.svg'), format='svg')
            plt.savefig(os.path.join(Sicherungsordner, f'{BlechNummer}_Plot_{Kraft} Kräfte Moving_Average.png'), format='png')
            
            plt.close()
```
