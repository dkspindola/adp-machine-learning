```python
# Anpassen der Kraftsensoren auf Bezeichnung der im Versuch aufgenommenen Kräfte --> Durch unterschiedliches Anschließen haben sich diese verändert 

# Vergleich von 4m Bezeichnung mit 2m Bezeichnungen
# 1-OW-OS Lateral Force = 4-UW-OS Lateral Force
# 1-OW-OS Axial Force = -4-UW-OS Axial Force
# 2-OW-RS Lateral Force = -3-UW-RS Lateral Force
# 2-OW-RS Axial Force= 3-UW-RS Axial Force
# 3-UW-RS Lateral Force = -2-OW-RS Lateral Force
# 3-UW-RS Axial Force = 2-OW-RS Axial Force
# 4-UW-OS Lateral Force = 1-OW-OS Lateral Force
# 4-UW-OS Axial Force = -1-OW-OS Axial Force

# Umbennen der Columns zur Vergleichbarkeit für die zentrierten Kräfte ohne Einlauf
df1.rename(columns={'1-OW-OS Lateral Force':'4-UW-OS Lateral Force','1-OW-OS Axial Force':'4-UW-OS Axial Force',
                            '2-OW-RS Lateral Force':'3-UW-RS Lateral Force','2-OW-RS Axial Force':'3-UW-RS Axial Force',
                            '3-UW-RS Lateral Force':'2-OW-RS Lateral Force','3-UW-RS Axial Force':'2-OW-RS Axial Force',
                            '4-UW-OS Lateral Force':'1-OW-OS Lateral Force','4-UW-OS Axial Force':'1-OW-OS Axial Force'
                            }, inplace=True)
# Vorzeichen ändern
df1['1-OW-OS Axial Force'] = df1['1-OW-OS Axial Force']*(-1)
df1['2-OW-RS Lateral Force'] =df1['2-OW-RS Lateral Force']*(-1)
df1['3-UW-RS Lateral Force'] = df1['3-UW-RS Lateral Force']*(-1)
df1['4-UW-OS Axial Force']= df1['4-UW-OS Axial Force']*(-1)

# Umbennen der Columns zur Vergleichbarkeit für die vorverarbeiteten Daten
df.rename(columns={'1-OW-OS Lateral Force':'4-UW-OS Lateral Force','1-OW-OS Axial Force':'4-UW-OS Axial Force',
                            '2-OW-RS Lateral Force':'3-UW-RS Lateral Force','2-OW-RS Axial Force':'3-UW-RS Axial Force',
                            '3-UW-RS Lateral Force':'2-OW-RS Lateral Force','3-UW-RS Axial Force':'2-OW-RS Axial Force',
                            '4-UW-OS Lateral Force':'1-OW-OS Lateral Force','4-UW-OS Axial Force':'1-OW-OS Axial Force'
                            }, inplace=True)
# Vorzeichen ändern
df['1-OW-OS Axial Force'] = df['1-OW-OS Axial Force']*(-1)
df['2-OW-RS Lateral Force'] =df['2-OW-RS Lateral Force']*(-1)
df['3-UW-RS Lateral Force'] = df['3-UW-RS Lateral Force']*(-1)
df['4-UW-OS Axial Force']= df['4-UW-OS Axial Force']*(-1)

# Umbennen der Columns zur Vergleichbarkeit für die Rohdaten
daten.rename(columns={'1-OW-OS Lateral Force':'4-UW-OS Lateral Force','1-OW-OS Axial Force':'4-UW-OS Axial Force',
                            '2-OW-RS Lateral Force':'3-UW-RS Lateral Force','2-OW-RS Axial Force':'3-UW-RS Axial Force',
                            '3-UW-RS Lateral Force':'2-OW-RS Lateral Force','3-UW-RS Axial Force':'2-OW-RS Axial Force',
                            '4-UW-OS Lateral Force':'1-OW-OS Lateral Force','4-UW-OS Axial Force':'1-OW-OS Axial Force'
                            }, inplace=True)
# Vorzeichen ändern
daten['1-OW-OS Axial Force'] = daten['1-OW-OS Axial Force']*(-1)
daten['2-OW-RS Lateral Force'] =daten['2-OW-RS Lateral Force']*(-1)
daten['3-UW-RS Lateral Force'] = daten['3-UW-RS Lateral Force']*(-1)
daten['4-UW-OS Axial Force']= daten['4-UW-OS Axial Force']*(-1)

Reihenfolge= ['1-OW-OS Lateral Force','1-OW-OS Axial Force','2-OW-RS Lateral Force', '2-OW-RS Axial Force','3-UW-RS Lateral Force','3-UW-RS Axial Force','4-UW-OS Lateral Force','4-UW-OS Axial Force']

#print(daten)
print(df)
df = df.reindex(columns=Reihenfolge)
df1 = df1.reindex(columns=Reihenfolge)
daten = daten.reindex(columns=Reihenfolge)
#print(df)
#print(daten)
print(df)
```
